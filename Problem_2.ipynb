```python
from tensorflow.keras.layers import Input,Dense
from tensorflow.keras.models import Model
def main():
  inputs=Input((2,))
  h1=Dense(4,activation='relu',name="Hidden layer 1")(inputs)
  h2=Dense(4,activation='relu',name="Hidden layer 2")(h1)
  outputs=Dense(2,activation='softmax',name="Output layer")(h2)
  model=Model(inputs,outputs)
  model.summary(show_trainable=True)
if __name__=='__main__':
  main()
```
